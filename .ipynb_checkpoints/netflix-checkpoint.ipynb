```python
all_cast = [val.strip() for sublist in df.cast.dropna().astype(str).str.split(",").tolist() for val in sublist]
unique_cast = list(set(all_cast))
print("{} \n".format(len(unique_cast)))
print(*unique_cast,sep="\n")
```

```python
df['duration']
```

```python
x = [str(i).split() for i in df['duration']]
```

```python
duration_value = []
duration_unit = []
for i in x:
    duration_value.append(i[0])
    if len(i) == 1:
        i.append("null")
    duration_unit.append(i[1])
```

```python
duration_unit = pd.Series(duration_unit)
duration_unit.value_counts()
# 3 null are movies
```

```python
duration_value = pd.Series(duration_value)
duration_value.value_counts()
```

```python
df["duration_units"] = duration_unit
df["duration_value"] = duration_value
```

```python
df.loc[df['duration_units']=="null",:]
```

```python
for id in [5541,5794,5813]:
    m = df.iloc[id,8]
    m = str(m).split()
    df.iloc[id,12] = m[1]
    df.iloc[id,13] = m[0]
    df.iloc[id,8] = "TV_MA"
```
